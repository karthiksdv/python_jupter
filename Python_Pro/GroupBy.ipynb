## Plot number of gold medals won by USA men and women through out the history


```python
umc = oo[(oo.NOC == 'USA') & (oo.Medal == 'Gold') ]
umc
umc.groupby(['Edition','Gender']).size().unstack('Gender',fill_value=0).plot(kind='line')
```

## Plot 5 athletes who won more gold medals

```python
g = oo.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head().plot(kind='bar');
```
