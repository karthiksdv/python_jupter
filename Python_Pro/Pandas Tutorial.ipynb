### Converters -- Adding value to the source data(excel)

```python
import pandas as pd

def convert_people_cell(cell):
    if cell == 'n.a.':
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell == 'not available':
        return 'None'
    return cell

df = pd.read_excel('stock_data.xlsx','Sheet1', converters = {'people': convert_people_cell,
                                                            'eps': convert_eps_cell})
df
```

# Merge DataFrames

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'city':['new yord','chicago','orlando'],
    'temperatutre':[21,14,35]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['chicago','new yord','orlando'],
    'windspeed':[12,9,14]
})
df2
```

```python
## Automatically match the data using CITY values like sql on query
df3 = pd.merge(df1,df2,on='city')
```

```python
df3
```

## Using union, intersections, inner & outer options

```python
df1 = pd.DataFrame({
    'city':['new yord','chicago','san fransisco'],
    'temperatutre':[21,14,35]
})

df2 = pd.DataFrame({
    'city':['chicago','new yord','baltimore','Dallas'],
    'windspeed':[12,9,8,'19']
})
```

```python
## Using inner and intersection. By default the how is inner
df3 = pd.merge(df1,df2,on='city',how='inner')
df3
```

```python
## Using outer and UNION.
df3 = pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
## Using left join
df3 = pd.merge(df1,df2,on='city',how='left')
df3
```

```python
## Using right join
df3 = pd.merge(df1,df2,on='city',how='right')
df3
```

### Using indicator flag which means the data from which side table

```python
## Using outer join and indicator flag
df3 = pd.merge(df1,df2,on='city',how='outer', indicator=True)
df3
```

## Suffixes post merge the data frames

```python
df1 = pd.DataFrame({
    'city':['new york','chicago','san fransisco'],
    'windspeed':[21,14,35],
    'humidity':[70,55,68]
})

df2 = pd.DataFrame({
    'city':['chicago','new york','baltimore','Dallas'],
    'windspeed':[12,9,8,19],
    'humidity':[84,64,57,79]
})
```

```python
### Here X and Y are default values.
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
### Customize the suffics values.
df3 = pd.merge(df1,df2,on='city', suffixes=('_left','_right'))
df3
```
