## Series
- Pandas Series is a one-dimensional labeled array capable of holding any data type.
- In simple terms, Pandas Series is nothing but a column in an excel sheet

```python
## Creating a python list
number = [1,2,3,4,5]
print(number)
print(type(number))
```

```python
## creating a Series
series_ex1 = pd.Series(['one','two','three','four','five'])
print(series_ex1)
print(type(series_ex1))
```

```python
## Converting a python list to series
number_series = pd.Series(number)
print(number_series)
print(type(number_series))
```

```python
## Create a index for the numbers
number_series.index = series_ex1
print(type(number_series))
number_series
```

```python
## Print the specific element using index label
print(number_series['one'])

## Print the specific element using index number(default)
print(number_series[1])
```

```python
## Series Multiple element selection
print(number_series[['one','four','five']])
```

```python
## Slicing the series
print(number_series[:2])  
print(number_series[3:])  
print(number_series[[0,3,4]])  
```

```python
## Conditional select in series
number_series[number_series > 2]
```

```python
number_series[number_series <= 2]
```

```python
number_series[number_series == 5]
```
